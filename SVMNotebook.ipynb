```python
import numpy as np
import matplotlib.pyplot as pl
%matplotlib inline



def plotSpectrum(y,Fs, color):
     """
     Plots a Single-Sided Amplitude Spectrum of y(t)
     """
     n = int(len(y)) # length of the signal
     k = np.arange(n)
     T = n/Fs
     frq = k/T # two sides frequency range
     frq = 50*frq[range(int(n/2))] # one side frequency range
     frq[0]=0
     Y = np.fft.fft(y)/n # fft computing and normalization
     Y = Y[range(int(n/2))]

#      print(frq,abs(Y))
     pl.plot(frq[:5000],2*abs(Y[:5000]),color) # plotting the spectrum
     pl.xlabel('f (Hz)')
     pl.ylabel('|Mag(f)|')
    
    
def plotSpectrumzoom(y,Fs, color,fend):
     """
     Plots a Single-Sided Amplitude Spectrum of y(t)
     """
     n = int(len(y)) # length of the signal
     k = np.arange(n)
     T = n/Fs
     frq = k/T # two sides frequency range
     frq = 50*frq[range(int(n/2))] # one side frequency range
     frq[0]=0
     Y = np.fft.fft(y)/n # fft computing and normalization
     Y = Y[range(int(n/2))]


     pl.plot(frq[:fend],2*abs(Y[:fend]),color) # plotting the spectrum
     pl.xlabel('f (Hz)')
     pl.ylabel('|Mag(f)|')
```

# Space Vector Modulation 

$$t_l = \frac{sin(\gamma)}{sin(\pi/3)}$$

```python
#Create a space vector reference
delta = np.arange(0,2*np.pi,0.001) #create an array from 0, 2pi in steps of 0.001
ts = np.arange(0,20e-3+1e-4,1e-4)
angn = 2*np.pi*50.0*ts
v =0.65

vsa = v*np.cos(angn) 
vsb = v*np.sin(angn)

pl.figure(101,figsize = (4,4))
pl.plot(vsa,vsb,'crimson')
pl.figure(102,figsize = (6,2.5))
pl.plot(angn,vsa,'r')
pl.xticks(np.linspace(angn[0],angn[-1],5))
# pl.plot(delta,vsb,'b')
pl.grid()
```

```python
#Angle within the sector
gamma = delta%(np.pi/3)
secnum = np.zeros(len(delta))
tl = (v/(2/3))*np.sin(gamma)/np.sin(np.pi/3)
for ii in range(len(delta)):
    secnum[ii] = int(delta[ii]/(np.pi/3))

pl.plot(delta,gamma, 'r')
pl.plot(delta,secnum,'navy')
pl.plot(delta,tl,'tomato')
pl.xticks(np.linspace(angn[0],angn[-1],5))
pl.show()
```

```python
ms = 0.6
vas = ms*0.5*np.sin(delta) + (1/6)*(ms)*0.5*np.sin(3*delta)
va = ms*0.5*np.sin(delta)

pl.figure(202, figsize = (6,2.5))
pl.plot(delta,va,'b')
pl.plot(delta,vas,'red')
pl.xticks(np.linspace(angn[0],angn[-1],5))
pl.show()
```
